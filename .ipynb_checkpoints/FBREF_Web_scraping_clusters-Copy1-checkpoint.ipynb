```python
al = get_player_data("https://fbref.com/en/players/9dbb75ca/Alexandre-Lacazette")
pea = get_player_data("https://fbref.com/en/players/d5dd5f1f/Pierre-Emerick-Aubameyang")
eh = get_player_data("https://fbref.com/en/players/1f44ac21/Erling-Haaland")
km = get_player_data("https://fbref.com/en/players/42fd9c7f/Kylian-Mbappe")
kb = get_player_data("https://fbref.com/en/players/70d74ece/Karim-Benzema")
tw = get_player_data("https://fbref.com/en/players/49fe9070/Timo-Werner")
rl = get_player_data("https://fbref.com/en/players/5eae500a/Romelu-Lukaku")
ec = get_player_data("https://fbref.com/en/players/527f063d/Edinson-Cavani")
```

```python
df = pd.concat([al, pea, eh, km, kb, tw, rl, ec])
```

```python
df = df[['name', 'non_penalty_goals', 'npx_g', 'shots_total', 'assists', 'x_a', 'npx_g_plus_x_a', 'shot_creating_actions', 'pass_completion_percent',
                                              'progressive_passes', 'progressive_carries', 'dribbles_completed', 'touches_att_pen']]
```

```python
df
```

```python
df_fix = pd.DataFrame()

df["non_penalty_goals_fix"] = pd.to_numeric(df["non_penalty_goals"])
df["npx_g_fix"] = pd.to_numeric(df["npx_g"])
df["shots_total_fix"] = pd.to_numeric(df["shots_total"])
df["assists_fix"] = pd.to_numeric(df["assists"])
df["x_a_fix"] = pd.to_numeric(df["x_a"])
df["npx_g_plus_x_a_fix"] = pd.to_numeric(df["npx_g_plus_x_a"])
df["shot_creating_actions_fix"] = pd.to_numeric(df["shot_creating_actions"])
df["progressive_passes_fix"] = pd.to_numeric(df["progressive_passes"])
df["progressive_carries_fix"] = pd.to_numeric(df["progressive_carries"])
df["dribbles_completed_fix"] = pd.to_numeric(df["dribbles_completed"])
df["touches_att_pen_fix"] = pd.to_numeric(df["touches_att_pen"])

df_fix["non_penalty_goals"] = df['non_penalty_goals_fix'][0]
df_fix["npx_g"] = df['npx_g_fix'][0]
df_fix["shots_total"] = df['shots_total_fix'][0]
df_fix["assists"] = df['assists_fix'][0]
df_fix["x_a"] = df['x_a_fix'][0]
df_fix["npx_g_plus_x_a"] = df['npx_g_plus_x_a_fix'][0]
df_fix["shot_creating_actions"] = df['shot_creating_actions_fix'][0]
df_fix["progressive_passes"] = df['progressive_passes_fix'][0]
df_fix["progressive_carries"] = df['progressive_carries_fix'][0]
df_fix["dribbles_completed"] = df['progressive_carries_fix'][0]
df_fix["touches_att_pen"] = df['touches_att_pen_fix'][0]
df_fix.head()
```

```python
KMeans_mod = KMeans(n_clusters=3, random_state=100)
```

```python
KMeans_mod.fit(df_fix)
```

```python
df_fix
```

```python
df_fix["Cluster"] = KMeans_mod.labels_
df_fix.groupby('Cluster').agg({
    'non_penalty_goals':'mean',
    'npx_g':'mean',
    'shots_total':'mean',
'assists':'mean',
'x_a':'mean',
'npx_g_plus_x_a':'mean',
'shot_creating_actions':'mean',
'progressive_passes':'mean',
'progressive_carries':'mean',
'dribbles_completed':'mean',
'touches_att_pen':'mean',
}).round(2)
```

```python
df_fix["Cluster"] = KMeans_mod.labels_
```

```python
df_fix['name'] = df.name
```
